```python
train=pd.get_dummies(train,columns=["P53"])
```

```python
train=train.drop(["P53_False"],axis=1)
```

```python
test=pd.get_dummies(test,columns=["P53"])
```

```python
test = test.drop(["P53_False"],axis=1)
```

```python
test
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import f1_score
from sklearn.metrics import precision_recall_curve
```

```python
x_train = train.drop(["molecular_subtype","id"],axis=1)
y_train = train["molecular_subtype"]
x_test = test.drop(["id"],axis=1)
```

```python
x_test
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier(criterion='entropy') #建立决策树对象
dtc.fit(x_train, y_train) #决策树拟合
y_test_pre = dtc.predict(x_test) #预测
print(y_test_pre)
```
